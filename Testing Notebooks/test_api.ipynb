```python
# Endpoint for generating the unified schema
schema_url = f"{url}/schema"

# Query parameters 
params = {
    "source": "databricks",
    "catalog_name": "derived_data_catalog",
    "schema_name": "dev_tvi_us_dashboards",
    "table_names": selected_tables,  
}

# Call the /schema endpoint
response = requests.get(schema_url, params=params)

if response.ok:
    # Extract the schema dictionary from the response
    result = response.json().get("schema", {})
    
    # Build the content for the Python file using triple quotes for preserving newlines
    python_schema = "schema = {\n"
    for table, columns in result.items():
        python_schema += f'    "{table}": """\n{columns}""",\n'
    python_schema += "}\n"
    
    # Write the content to schema.py
    with open("schema_tvi.py", "w") as f:
        f.write(python_schema)
    print("Schema saved to schema.py")
else:
    print(f"Error retrieving schema: {response.status_code} - {response.text}")
```

```python
import requests
import json

# Base URL for the API host
url = "http://localhost:8084"  # adjust as needed
schema_url = f"{url}/schema"

# Your list of table names (as file keys in S3 or table names in Databricks)
# table_names = ["report_segment_loads", "another_table", "yet_another_table"]
table_names = ["concentrated_data_cube"]

# Parameters for the request (adjust source and credentials accordingly)
params = {
    "source": "databricks",
    "catalog_name": "data_management_dev",
    "schema_name": "concentrated_data_cube",
    "table_names": table_names  # FastAPI expects multiple query parameters for lists
}

# Call the /schema endpoint
response = requests.get(schema_url, params=params)

if response.ok:
    # The endpoint returns a dictionary under the "schema" key.
    result = response.json().get("schema", {})
    
    # Build a string that starts with "schema = " and contains the unified schema.
    # For each table, we use a triple-quoted string so that newlines are preserved.
    schema_str = "schema = {\n"
    for table, schema_text in result.items():
        # Optionally, indent the schema text for better readability.
        # Replace any existing newlines with newlines plus some indent (e.g., 4 spaces).
        indented_text = schema_text.replace("\n", "\n    ")
        schema_str += f"    '{table}': '''{indented_text}''',\n"
    schema_str += "}\n"
    
    # Print the formatted schema variable assignment
    print(schema_str)
    
    # Save the schema string locally (this file is no longer pure JSON; it's a Python assignment)
    with open("local_schema.txt", "w") as f:
        f.write(schema_str)
    print("Schema saved to local_schema.txt")
else:
    print(f"Error retrieving schema: {response.status_code} - {response.text}")
```

```python
import os
import json

# Initialize the summaries dictionary
summaries = {}

# Iterate through all files in the folder
for file_name in os.listdir("json_summaries_adapt"):
    if file_name.endswith(".json"):  # Check if the file is a JSON file
        table_name = file_name.split(".json")[0]  # Extract the table name (file name without .json)
        file_path = os.path.join("json_summaries_adapt", file_name)  # Construct the full file path
        with open(file_path, "r", encoding="utf-8") as f:
            summaries[table_name] = json.load(f)  # Read and store the JSON content

# Print the summaries dictionary to verify
print(summaries)
```

```python
# Iterate over each summary and store it in the vector database
for table, summary in summaries.items():
    payload = {
        "enriched_text": "",  
        "json_summary": summary,
        "db_path": "./vector_db_adapt",  
        "collection_name": "adapt"  
    }
    store_resp = requests.post(f"{url}/vectorstore/store", json=payload)
    if store_resp.ok:
        print(f"Vector store update for table '{table}':")
        print(store_resp.json())
    else:
        print(f"Error storing table '{table}':", store_resp.text)
```

```python
for s3_folder in s3_paths:
    print(f"\nProcessing S3 folder: {s3_folder}")
    
    # Step 1: List valid files in the folder using the /tables endpoint.
    params_tables = {
        "source": "s3",
        "s3_folder_path": s3_folder
    }
    resp_tables = requests.get(f"{url}/tables", params=params_tables)
    if resp_tables.ok:
        data_tables = resp_tables.json()
        valid_files = data_tables.get("tables", [])
        if not valid_files:
            print(f"  No valid CSV/Parquet files found in {s3_folder}.")
            continue
        # Choose one file 
        sample_key = valid_files[0]
        print(f"  Selected file: {sample_key}")
    else:
        print(f"  Error listing files for {s3_folder}: {resp_tables.text}")
        continue

    # Summarize the selected file using the /summarizer/summarize endpoint
    params_summarize = {
        "source": "s3",
        "s3_folder_path": s3_folder,
        "file_key": sample_key,
        "n_samples": 3,
        "use_llm": True
    }
    resp_summary = requests.get(f"{url}/summarizer/summarize", params=params_summarize)
    if resp_summary.ok:
        data_summary = resp_summary.json()
        summary = data_summary.get("summary", {})
        print("  Summary generated for file.")
    else:
        print(f"  Error summarizing file {sample_key}: {resp_summary.text}")
        continue

    # Store the summary in the vector DB under the collection "ninjacat".
    payload_store = {
        "enriched_text": "",  
        "json_summary": summary,
        "db_path": "./chroma_db1",
        "collection_name": "ninjacat"  
    }
    resp_store = requests.post(f"{url}/vectorstore/store", json=payload_store)
    if resp_store.ok:
        print(f"  Stored summary for {sample_key} in vector DB under collection 'ninjacat'.")
    else:
        print(f"  Error storing summary for {sample_key}: {resp_store.text}")
```

```python
# Input user query
user_query = input("Enter your query: ")

# Validate and parse the query
if not user_query.startswith("@"):
    print("Error: Query must start with '@schema_name'.")
else:
    tokens = user_query.split()
    if len(tokens) < 2:
        print("Invalid query format. Please use '@schema_name your_query'.")
    else:
        # Extract schema name 
        schema_token = tokens[0]
        schema_name = schema_token[1:]
        actual_query = " ".join(tokens[1:])

        print(f"Activating the {schema_name} agent...")
        
        time.sleep(15)  # Wait for 15 secs
        print(f"Finding best table for you in {schema_name}...")

        # Set up the retrieval parameters
        retrieval_params = {
            "query": actual_query,
            "k": 3,
            "db_path": "./chroma_db1",
            "collection_name": schema_name  
        }
        
        # Call the retrieval endpoint
        retrieval_response = requests.get(f"{url}/retriever/retrieve", params=retrieval_params)
        
        if retrieval_response.ok:
            retrieval_result = retrieval_response.json()
            best_table = retrieval_result.get("best_table_name", "UnknownTable")
            print(f"Found best table for your query: {best_table}")
        else:
            print("Error in retrieval:", retrieval_response.text)
```

```python
# Input user query
user_query = input("Enter your query: ")

# Validate and parse the query
if not user_query.startswith("@"):
    print("Error: Query must start with '@collection_name'.")
else:
    tokens = user_query.split()
    if len(tokens) < 2:
        print("Invalid query format. Please use '@collection_name your_query'.")
    else:
        # Extract collection name and actual query text.
        collection_token = tokens[0]
        collection_name = collection_token[1:]  # Remove the '@'
        # Allowed collections
        allowed_collections = ["ninjacat", "dev_tvi_us_dashboards"]
        if collection_name.lower() not in allowed_collections:
            print(f"Error: Collection must be one of {allowed_collections}.")
        else:
            actual_query = " ".join(tokens[1:])
            
            # Print interactive messages.
            print(f"Activating the {collection_name} agent...")
            time.sleep(15)  # Wait for 15 seconds.
            print(f"Finding best table for you in {collection_name}...")
            
            # Set up the retrieval parameters.
            retrieval_params = {
                "query": actual_query,
                "k": 3,
                "db_path": "./chroma_db1",
                "collection_name": collection_name.lower()  
            }
            
            retrieval_response = requests.get(f"{url}/retriever/retrieve", params=retrieval_params)
            
            if retrieval_response.ok:
                retrieval_result = retrieval_response.json()
                best_table = retrieval_result.get("best_table_name", "UnknownTable")
                print(f"Found best table for your query: {best_table}")
            else:
                print("Error in retrieval:", retrieval_response.text)
```
